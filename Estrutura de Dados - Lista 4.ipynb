```python
    df = pd.DataFrame(table[1:], columns=table[0])
    
    df = df[['Estado', 'Mulheres vítimas de homicídio em 2017', 'População do Estado em 2017', 'Mulheres vítimas de homicídio em 2018', 'População do Estado em 2018', 'Variação (em %)']]
    
    df = df.apply(pd.to_numeric, errors='ignore')
    df.head(28)
    
    
```

```python
trace1 = go.Bar(
    x = df['Estado'],
    y = df['Mulheres vítimas de homicídio em 2017'],
    name = '2017',
    )
trace2 = go.Bar(
    x = df['Estado'],            
    y = df['Mulheres vítimas de homicídio em 2018'],
    name = '2018',
)


layout = go.Layout(
    title = '<b> Comparação de homicídio nos anos de 2017 e 2018',
    titlefont = {
        'family':'Arial',
        'size':24
    }
)
data = [trace1, trace2]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)

##percebe-se que as regiões em que há mais homícidios no Brasil, se trata de São Paulo, Bahia, Ceará, Rio de Janeiro
##e Minas Gerais
```

```python
statesArr = df['Estado']
pop17Arr = df['População do Estado em 2017']
die17Arr = df['Mulheres vítimas de homicídio em 2017']

popEstados2017 = go.Bar(
    x = [statesArr[24], statesArr[4], statesArr[5], statesArr[18], statesArr[12]],
    y = [pop17Arr[24], pop17Arr[4], pop17Arr[5], pop17Arr[18], pop17Arr[12]],
    name = 'População',
    marker={'color':'#000000'}
)

diesEstados2017 = go.Bar(
    x = [statesArr[24], statesArr[4], statesArr[5], statesArr[18], statesArr[12]],
    y = [die17Arr[24], die17Arr[4], die17Arr[5], die17Arr[18], die17Arr[12]],
    name = 'Homicídios',
    marker={'color':'#ff0000'}
)


layout = go.Layout(
    title = '<b> Comparação de homicídio de mulheres em 2017',
    titlefont = {
        'family':'Arial',
        'size':24
    },
    barmode='stack'
)
data = [diesEstados2017, popEstados2017]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
